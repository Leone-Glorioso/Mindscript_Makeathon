```python
import gradio as gr
import json

# 🌐 Avatar URLs
USER_AVATAR = "https://img.freepik.com/premium-vector/avatar-icon0002_750950-43.jpg"
BOT_AVATAR = "https://encrypted-tbn0.gstatic.com/images?q=tbn:ANd9GcSMzp4cfvB5JfHaMgMzcJSlDSoSj3g0CDQqIA&s"

# 💬 Styling
USER_COLOR = "#15314b"
BOT_COLOR = "#1e3a5f"

# # 🗃️ Store history
conversation_histories = {"default": []}

# 🧠 Actual chat logic (cleaning assistant's prefix)
def gradio_chat(message, history):
    print(message)
    response, _ = insurance_chat(message)
    print(response)
    return response.replace("<|eot_id|>assistant", "", 1)

# 💬 Bubble formatter
def format_message(message, is_user):
    avatar = USER_AVATAR if is_user else BOT_AVATAR
    name = "Client" if is_user else "Michelle The Assistant"
    bg_color = USER_COLOR if is_user else BOT_COLOR
    direction = "row-reverse" if is_user else "row"
    text_align = "right" if is_user else "left"

    return f"""
    <div style="display: flex; flex-direction: {direction}; align-items: flex-start; width: 100%; margin: 15px 0;">
        <img src="{avatar}" width="48" height="48" style="border-radius: 50%; margin: 0 12px;">
        <div style="width: 500px;">
            <div style="font-size: 0.85em; color: #aaa; text-align: {text_align}; margin-bottom: 3px;">{name}</div>
            <div style="background: {bg_color}; color: white; padding: 12px 16px; border-radius: 15px; word-wrap: break-word;">
                {message}
            </div>
        </div>
    </div>
    """

# ⏳ Loading animation
def typing_indicator():
    return f"""
    <div style=\"display: flex; align-items: center; gap: 12px; width: 500px; margin-left: 0;\">
        <img src=\"{BOT_AVATAR}\" width=\"48\" height=\"48\" style=\"border-radius: 50%;\">
        <div class=\"typing-indicator\" style=\"display: inline-flex; gap: 4px;\">
            <span></span><span></span><span></span>
        </div>
    </div>
    """

# 🔁 Handle submission
# 🔁 First: Handle user submission and show typing bubble
def user_submit(user_message, chat_display):
    user_msg_html = format_message(user_message, is_user=True)
    typing_html = typing_indicator()

    interim_display = chat_display + user_msg_html + typing_html

    return "", interim_display, user_message

# 💬 Second: Generate response and update bubble
def bot_respond(user_message, chat_display):
    # Use the actual message history
    response = gradio_chat(user_message, conversation_histories["default"])
    bot_msg_html = format_message(response, is_user=False)

    conversation_histories["default"].append(user_message)
    conversation_histories["default"].append(response)

    cleaned_display = chat_display.rsplit(typing_indicator(), 1)[0]
    final_display = cleaned_display + bot_msg_html
    return final_display



# 🧹 Reset
def reset_history():
    conversation_histories["default"] = []
    greeting = "Hello, my name is Michelle. How can I help you today?"
    conversation_histories["default"].append(greeting)
    return "", format_message(greeting, is_user=False)

# 🎨 UI
with gr.Blocks(title="Sleek Insurance Chatbot", css="""
    #send-btn, #reset-btn {
        background-color: #1e3a5f;
        color: white;
        border-radius: 12px;
        padding: 10px 16px;
        font-weight: bold;
        transition: 0.2s ease-in-out;
        border: none;
        min-width: 100px;
    }
    #send-btn:hover, #reset-btn:hover {
        background-color: #28527a;
    }
    .gr-textbox input {
        border-radius: 12px !important;
        padding: 12px;
        border: 1px solid #ccc;
        font-size: 1rem;
    }
    .chat-header {
        display: flex;
        align-items: center;
        gap: 12px;
        padding-bottom: 10px;
        border-bottom: 1px solid #ccc;
        margin-bottom: 20px;
    }
    .chat-header img {
        width: 50px;
        height: 50px;
        border-radius: 50%;
    }
    .chat-header .title {
        font-weight: bold;
        font-size: 1.3rem;
        margin: 0;
        color: #1e3a5f;
    }
    .chat-header .subtitle {
        font-size: 0.9rem;
        color: #666;
        margin: 0;
    }
    .gr-overlay-background {
        display: none !important;
        background: none !important;
    }
    .typing-indicator {
        display: inline-flex;
        gap: 4px;
        margin-top: 6px;
    }
    .typing-indicator span {
        width: 10px;
        height: 10px;
        background-color: #bbb;
        border-radius: 50%;
        animation: bounce 1.4s infinite ease-in-out both;
    }
    .typing-indicator span:nth-child(1) { animation-delay: -0.32s; }
    .typing-indicator span:nth-child(2) { animation-delay: -0.16s; }
    .typing-indicator span:nth-child(3) { animation-delay: 0s; }

    @keyframes bounce {
        0%, 80%, 100% { transform: scale(0); }
        40% { transform: scale(1); }
    }
""") as demo:

    gr.HTML(
        """
        <div class="chat-header">
            <img src="https://encrypted-tbn0.gstatic.com/images?q=tbn:ANd9GcSMzp4cfvB5JfHaMgMzcJSlDSoSj3g0CDQqIA&s" alt="Bot">
            <div>
                <p class="title">Hellas Direct AI Assistant</p>
                <p class="subtitle">Here to help you in your emergency</p>
            </div>
        </div>
        """
    )

    chat_display = gr.HTML()
    user_input = gr.Textbox(placeholder="Ask me anything about your insurance...", show_label=False, scale=9, elem_classes=["gr-textbox"])
    submit_btn = gr.Button("Submit", elem_id="send-btn", scale=1)
    reset_btn = gr.Button("New Chat", elem_id="reset-btn")

    # Initial greeting
    demo.load(fn=reset_history, outputs=[user_input, chat_display])

    msg_state = gr.State()
    
    submit_btn.click(
        user_submit,
        [user_input, chat_display],
        [user_input, chat_display, msg_state],
        show_progress=False
    ).then(
        bot_respond,
        [msg_state, chat_display],  # ✅ Use State instead
        [chat_display],
        show_progress=False
    )
    
    user_input.submit(
        user_submit,
        [user_input, chat_display],
        [user_input, chat_display, msg_state],
        show_progress=False
    ).then(
        bot_respond,
        [msg_state, chat_display],  # ✅ again, use msg_state not textbox
        [chat_display],
        show_progress=False
    )

    reset_btn.click(reset_history, outputs=[user_input, chat_display])

demo.launch()
```
